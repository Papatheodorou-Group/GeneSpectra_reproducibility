```r
library(tidyverse)
library(biomaRt)
```

```r
tbl <- read_csv("zebrafish_gene_name_ct_final.csv")
```

```r
tbl$gene...1
```

```r
ensembl <- useEnsembl(biomart = "ensembl", 
                   dataset = "drerio_gene_ensembl", 
                   mirror = "useast",
                      version = 80)
```

```r
ensembl
```

```r
attributes = listAttributes(ensembl)
head(attributes, n=20)
```

```r
res=getBM(attributes = c('ensembl_gene_id', 'ensembl_peptide_id', "external_gene_name"),
      filters = 'external_gene_name',
      values = tbl$gene...1, 
      mart = ensembl)
```

```r
res %>% filter(external_gene_name %in% tbl$gene...1) %>% write_csv("drerio_GRCz10_ensemblr80_gene_pep_name.csv")
```

```r
ensembl <- useEnsembl(biomart = "ensembl", 
                   dataset = "xtropicalis_gene_ensembl", 
                   mirror = "useast",
                      version = 106)
```

```r
ensembl
```

```r
attributes = listAttributes(ensembl)
head(attributes, n=20)
```

```r
res2=getBM(attributes = c('ensembl_gene_id', 'ensembl_peptide_id', "external_gene_name"),
      filters = 'external_gene_name',
      values = tbl2$gene...1, 
      mart = ensembl)
```

```r
res2
```

```r
res2 %>% filter(external_gene_name %in% tbl2$gene...1) %>%  write_csv("xtropicalis_v9.1_ensemblr106_gene_pep_name.csv")
```
