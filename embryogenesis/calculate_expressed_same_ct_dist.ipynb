```r
species_1 <- 'drerio'
species_2 <- 'xtropicalis'
```

```r
tbl <- read_csv(paste0("../homology_mapped_o2o_", species_1, "_", species_2, "_1TPM.csv"))
```

```r
intersection_length <- function(list1, list2) {
  length(intersect(list1, list2))
}
```

```r
union_length <- function(list1, list2) {
  length(union(list1, list2))
}
```

```r
tbl = tbl %>% select(eval(paste0(species_2, "_homolog_associated_gene_name")), 
                     eval(paste0("gene_", species_1)), 
                     eval(paste0("groups_expressed_", species_2)), 
                     eval(paste0("groups_expressed_", species_1)), 
                     eval(paste0("dist_category_", species_2)), 
                     eval(paste0("dist_category_", species_1))) %>% 
mutate(groups_expressed_sp1 = strsplit(get(paste0("groups_expressed_", species_1)), split = ';'))%>% 
mutate(groups_expressed_sp2 = strsplit(get(paste0("groups_expressed_", species_2)), split = ';'))
```

```r
head(tbl)
```

```r
tbl %>%
  rowwise() %>%
  mutate(n_intersect = intersection_length(groups_expressed_sp1, groups_expressed_sp2))%>%
  mutate(n_union = union_length(groups_expressed_sp1, groups_expressed_sp2)) %>% 
mutate(n_diff_ct = n_union - n_intersect) %>% 
filter(get(paste0("dist_category_", species_2)) == get(paste0("dist_category_", species_1))) %>% 
filter(get(paste0("dist_category_", species_2)) == 'expressed in over 90%') %>% 
ungroup() %>% 
mutate(sum_same_ct = sum(n_intersect))%>% 
mutate(sum_diff_ct = sum(n_diff_ct)) %>% 
select(sum_same_ct, sum_diff_ct) %>% 
unique() %>% 
mutate(species_1 = species_1, species_2 = species_2, dist_category_sp1 = 'expressed in over 90%', dist_category_sp2 = 'expressed in over 90%')
```

```r
tbl <- read_csv(paste0("../homology_mapped_o2o_", species_1, "_", species_2, "_1TPM.csv"))
    tbl = tbl %>% select(eval(paste0(species_2, "_homolog_associated_gene_name")), 
                     eval(paste0("gene_", species_1)), 
                     eval(paste0("groups_expressed_", species_2)), 
                     eval(paste0("groups_expressed_", species_1)), 
                     eval(paste0("dist_category_", species_2)), 
                     eval(paste0("dist_category_", species_1))) %>% 
mutate(groups_expressed_sp1 = strsplit(get(paste0("groups_expressed_", species_1)), split = ';'))%>% 
mutate(groups_expressed_sp2 = strsplit(get(paste0("groups_expressed_", species_2)), split = ';'))

    
add = tbl %>%
  rowwise() %>%
  mutate(n_intersect = intersection_length(groups_expressed_sp1, groups_expressed_sp2))%>%
  mutate(n_union = union_length(groups_expressed_sp1, groups_expressed_sp2)) %>% 
mutate(n_diff_ct = n_union - n_intersect) %>% 
filter(get(paste0("dist_category_", species_2)) == get(paste0("dist_category_", species_1))) %>% 
filter(get(paste0("dist_category_", species_2)) == 'expressed in over 90%') %>% 
ungroup() %>% 
mutate(sum_same_ct = sum(n_intersect))%>% 
mutate(sum_diff_ct = sum(n_diff_ct)) %>% 
select(sum_same_ct, sum_diff_ct) %>% 
unique() %>% 
mutate(species_1 = species_1, species_2 = species_2, dist_category_sp1 = 'expressed in over 90%', dist_category_sp2 = 'expressed in over 90%')



pie = add %>% 
select(sum_same_ct, sum_diff_ct) %>% 
unique() %>% 
t() %>% as.data.frame() %>% 
rownames_to_column('kind') %>% 
mutate(kind = gsub("sum_diff_ct", "Different", kind)) %>% 
mutate(kind = gsub("sum_same_ct", "Same", kind)) %>% 
mutate(pct = V1 / sum(V1)) %>% 
ggplot(aes(x = factor(1), y = pct, fill = kind)) +
geom_bar(stat = 'identity', position='stack', color = 'black', size = 0.2) +
geom_text(aes(x = 1.6, label = scales::percent(pct, accuracy = .1), alpha = NULL), position = position_stack(vjust = .5), show.legend=FALSE) + 
coord_polar(theta = 'y') +
scale_fill_manual(name = 'Cell type', values = c("Different" = '#cccccc', "Same" = '#7a7a7a')) +
common_0x +
theme(axis.line.x = element_blank(),
      axis.text.y = element_blank(),
      axis.ticks.y = element_blank(),
      axis.text.x = element_blank(),
     axis.line.y = element_blank(),
     axis.title.x = element_blank(),
     axis.title.y = element_blank()) +
labs(title = '')

write_csv(add, "dist_all_species_90_same_ct.csv")    
saveRDS(pie, paste0(species_1, "_", species_2, "_dist_cell_type_pct_90_pie.rds"))
ggsave(plot = pie, filename = paste0(species_1, "_", species_2, "_dist_cell_type_pct_90_pie.pdf"), height = 3, width = 4)
    
```

```r
tbl <- read_csv(paste0("../homology_mapped_o2o_", species_1, "_", species_2, "_1TPM.csv"))
    tbl = tbl %>% select(eval(paste0(species_2, "_homolog_associated_gene_name")), 
                     eval(paste0("gene_", species_1)), 
                     eval(paste0("groups_expressed_", species_2)), 
                     eval(paste0("groups_expressed_", species_1)), 
                     eval(paste0("dist_category_", species_2)), 
                     eval(paste0("dist_category_", species_1))) %>% 
mutate(groups_expressed_sp1 = strsplit(get(paste0("groups_expressed_", species_1)), split = ';'))%>% 
mutate(groups_expressed_sp2 = strsplit(get(paste0("groups_expressed_", species_2)), split = ';'))

tbl = tbl %>% 
mutate(dist_category_sp2 = get(paste0("dist_category_", species_2)))%>% 
mutate(dist_category_sp1 = get(paste0("dist_category_", species_1))) %>% 
rowwise() %>%
mutate(n_intersect = intersection_length(groups_expressed_sp1, groups_expressed_sp2))%>%
mutate(n_union = union_length(groups_expressed_sp1, groups_expressed_sp2)) %>% 
mutate(n_diff_ct = n_union - n_intersect) %>% 
filter(dist_category_sp2 %in% c("expressed in over 30%","expressed in less than 30%")) %>% 
filter(dist_category_sp1 %in% c("expressed in over 30%","expressed in less than 30%")) %>% 
ungroup() %>% 
group_by(dist_category_sp2, dist_category_sp1) %>% 
mutate(sum_same_ct = sum(n_intersect))%>% 
mutate(sum_diff_ct = sum(n_diff_ct)) %>% 
select(sum_same_ct, sum_diff_ct) %>% 
unique() %>% 
mutate(species_1 = species_1, species_2 = species_2)


pie = tbl %>% 
mutate(category = case_when(
    dist_category_sp1 == dist_category_sp2 & dist_category_sp1 == 'expressed in less than 30%' ~'Stay < 30%',
    dist_category_sp1 == dist_category_sp2 & dist_category_sp1 == 'expressed in over 30%' ~'Stay > 30%',
    dist_category_sp1 != dist_category_sp2 & dist_category_sp2 == 'expressed in over 30%' ~'To > 30%',
    dist_category_sp1 != dist_category_sp2 & dist_category_sp2 == 'expressed in less than 30%' ~'To < 30%',
)) %>% 
pivot_longer(cols = c(sum_same_ct, sum_diff_ct), names_to = 'kind', values_to = 'count') %>% 
mutate(kind = gsub("sum_diff_ct", "Different", kind)) %>% 
mutate(kind = gsub("sum_same_ct", "Same", kind)) %>% 
ungroup() %>% 
mutate(pct = count / sum(count)) %>% 
ggplot(aes(x = factor(1), y = pct, fill = category, alpha = kind)) +
geom_bar(stat = 'identity', position='stack', color = 'black', size = 0.2) +
geom_text(aes(x = 1.6, label = scales::percent(pct, accuracy = .1), alpha = NULL), position = position_stack(vjust = .5), show.legend=FALSE) + 
coord_polar(theta = 'y') +
scale_fill_nejm(name = 'Cell type', alpha = 0.8) +
scale_alpha_discrete(name = 'Cell type', range = c(0.3, 0.8)) +
common_0x +
theme(axis.line.x = element_blank(),
      axis.text.y = element_blank(),
      axis.ticks.y = element_blank(),
      axis.text.x = element_blank(),
     axis.line.y = element_blank(),
     axis.title.x = element_blank(),
     axis.title.y = element_blank()) +
labs(title = '')

write_csv(add, "dist_all_species_30_same_ct.csv")       
saveRDS(pie, paste0(species_1, "_", species_2, "_dist_cell_type_pct_30_pie.rds"))
ggsave(plot = pie, filename = paste0(species_1, "_", species_2, "_dist_cell_type_pct_30_pie.pdf"), height = 3, width = 4)
    
```
