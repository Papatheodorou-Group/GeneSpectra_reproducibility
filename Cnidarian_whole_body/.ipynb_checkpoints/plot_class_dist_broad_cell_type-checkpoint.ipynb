```r
hvul <- read_csv("metacell_Hvul_geo_1e6_0.1lim_3fold_broad_type_classes.csv")
```

```r
xboc = read_csv("metacell_Xboc_geo_1e6_0.1lim_3fold_broad_type_classes.csv")
```

```r
xboc_genespectra = xboc %>% 
ggplot(aes(x = spec_category, fill = spec_category)) +
geom_histogram(stat = 'count') +theme_bw() +
labs(title = 'Xboc genespectrs classes, MARS-seq') + common_45x
```

```r
xboc_genespectra
```

```r
xboc_tau = xboc  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
ggplot(aes(x = tau)) + 
geom_histogram() + theme_bw()  +
labs(title = 'Xboc genespectrs classes, MARS-seq')
```

```r
spis_genespectra = spis %>% 
ggplot(aes(x = spec_category, fill = spec_category)) +
geom_histogram(stat = 'count') +theme_bw() +
labs(title = 'Spis genespectrs classes, MARS-seq') + common_45x
```

```r
spis_tau = spis  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
ggplot(aes(x = tau)) + 
geom_histogram() + theme_bw()  +
labs(title = 'Spis genespectrs classes, MARS-seq')
```

```r
spis  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
group_by(tau_category, spec_category) %>% count()
```

```r
hvul_genespectra = hvul %>% 
ggplot(aes(x = spec_category, fill = spec_category)) +
geom_histogram(stat = 'count') +theme_bw() +
labs(title = 'Hvul genespectrs classes, DROP-seq') + common_45x

hvul_tau = hvul  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
ggplot(aes(x = tau)) + 
geom_histogram() + theme_bw() +
labs(title = 'Hvul genespectrs classes, DROP-seq')
```

```r
nvec_genespectra = nvec %>% 
ggplot(aes(x = spec_category, fill = spec_category)) +
geom_histogram(stat = 'count') +theme_bw() +
labs(title = 'Nvec genespectrs classes, MARS-seq') + common_45x

nvec_tau = nvec  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
ggplot(aes(x = tau)) + 
geom_histogram() + theme_bw() +
labs(title = 'Nvec genespectrs classes, MARS-seq')


xesp_genespectra = xesp %>% 
ggplot(aes(x = spec_category, fill = spec_category)) +
geom_histogram(stat = 'count') +theme_bw() +
labs(title = 'Xesp genespectrs classes, 10X v3') + common_45x

xesp_tau = xesp  %>% 
filter(tau_category	!= 'lowly expressed') %>% 
ggplot(aes(x = tau)) + 
geom_histogram() + theme_bw() +
labs(title = 'Xesp genespectrs classes, 10X v3')
```

```r
ortho <- read_tsv("../orthologous_pairs.txt", col_names = c("gene1", "gene2"))
```

```r
nvec_hvul_ortho = ortho %>% filter(((gene1 %in% nvec[['gene...1']]) & (gene2 %in% hvul[['gene...1']])) | ((gene2 %in% nvec[['gene...1']]) & (gene1 %in% hvul[['gene...1']])))
```

```r
head(nvec_hvul_ortho)
```

```r
nvec %>% (gene...1	%in% nvec_hvul_ortho$gene1)
```

```r
tail(nvec_hvul_ortho)
```

```r
nvec_spis_ortho = ortho %>% filter(((gene1 %in% nvec[['gene...1']]) & (gene2 %in% spis[['gene...1']])) | ((gene2 %in% nvec[['gene...1']]) & (gene1 %in% spis[['gene...1']])))
```

```r
head(nvec_spis_ortho)
```

```r
xesp_spis_ortho = ortho %>% filter(((gene1 %in% xesp[['gene...1']]) & (gene2 %in% spis[['gene...1']])) | ((gene2 %in% xesp[['gene...1']]) & (gene1 %in% spis[['gene...1']])))
```

```r
head(xesp_spis_ortho)
```

```r
hvul_xesp_ortho = ortho %>% filter(((gene1 %in% xesp[['gene...1']]) & (gene2 %in% hvul[['gene...1']])) | ((gene2 %in% xesp[['gene...1']]) & (gene1 %in% hvul[['gene...1']])))
```

```r
head(hvul_xesp_ortho)
```

```r
ortho_all = merge(merge(nvec_spis_ortho, xesp_spis_ortho, by='gene2'), hvul_xesp_ortho, by.x = 'gene1.y', by.y = 'gene1')
```

```r
head(ortho_all)
```
